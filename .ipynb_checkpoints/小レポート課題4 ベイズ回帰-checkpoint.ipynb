```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from IPython.display import clear_output
import japanize_matplotlib
```

## $\alpha, \beta$を適当に定めてベイズ回帰

```python
class BayesLinearPolynomialRegression(object):
    def __init__(self, M=3, alpha=1e-2, beta=1):
        self.M = M
        #alpha, betaは入力引数から適当に決める
        self.alpha = alpha
        self.beta = beta
    
    #基底関数
    def polynomialize(self, x):
        phi = np.array([x**i for i in range(self.M + 1)]).T
        return phi

    def fit(self, X, t):
        #基底関数
        phi = self.polynomialize(X)
        #事後分布の分散
        self.w_s = np.linalg.inv(self.alpha * np.eye(phi.shape[1])
                                 + self.beta * phi.T.dot(phi))
        #事後分布の平均
        self.w_m = self.beta * self.w_s.dot(phi.T.dot(t))
    
    #予測分布 (今回は使わない)
    def predict(self, x):
        phi = self.polynomialize(x)
        t_m = phi.dot(self.w_m)
        #予測分布の分散
        t_v = 1/self.beta + np.sum(phi.dot(self.w_s) * phi, axis=1)
        #予測分布の標準偏差
        t_s = np.sqrt(t_v)
        return t_m, t_v
```

```python
def f(x):
    #return np.sin(2 * np.pi * x)
    return -100*x*((x-0.5)**2)*(x-1)

#データ生成
def generate_data(func, noise_scale, N):
    x = np.random.uniform(size=N)
    t = func(x) + np.random.normal(scale=noise_scale, size=x.shape)
    return x, t

#描画
def display_predict(truefunc,
                    X_train,
                    t_train,
                    X_test,
                    t_test_m,
                    t_test_s,
                    subindex,
                    text=None,
                    name=None,
                    show=True):
    clear_output(True)
    #fig = plt.figure(figsize=(8, 6))
    plt.subplot(1, 3, subindex)
    plt.fill_between(
        X_test,
        t_test_m - t_test_s,
        t_test_m + t_test_s,
        color='r',
        alpha=0.25)
    plt.plot(X_test, truefunc(X_test), label='Target Curve', c='blue')
    plt.scatter(
        X_train,
        t_train,
        facecolors='none',
        edgecolors='black',
        label='Observated Points',
        alpha=0.6)
    plt.plot(X_test, t_test_m, c='r', label='Predict Curve')
    if text is not None:
        plt.title(text, fontsize=12)
    plt.xlabel('$x$', fontsize=20)
    plt.ylabel('$y$', fontsize=20)
    plt.ylim((-2.5, 4.00))
    if name is not None:
        plt.savefig('{}.png'.format(name))
    if show:
        plt.legend()
        plt.show()


def display_predict_with_lines(truefunc,
                               X_train,
                               t_train,
                               model,
                               subindex=3,
                               linenum=3,
                               text=None,
                               name=None,
                               num=200):
    X_test = np.linspace(0, 1, 200)
    t_m, t_s = model.predict(X_test)
    display_predict(
        truefunc, X_train, t_train, X_test, t_m, t_s, subindex=subindex, text=text, show=False)
    w_m, w_s = model.w_m, model.w_s
    func = lambda x, w: sum([_w * x**i for i, _w in enumerate(w)])
    for _ in range(linenum):
        #事後分布からのサンプリング
        w_sample = multivariate_normal.rvs(w_m, w_s)
        plt.plot(
            X_test,
            func(X_test, w_sample),
            linestyle='--',
            c='purple',
            alpha=0.5,
            linewidth=1)
    plt.xlim((0.0, 1.0))
    if name is not None:
        plt.savefig('{}.png'.format(name))
    if subindex==1: plt.legend()
    plt.grid()
    plt.savefig("M_change.png")
```

## $\alpha, \beta$をエビデンス関数の最大化で最適化

```python
class BayesLinearPolynomialRegressionWithEvidence(
        BayesLinearPolynomialRegression):
    def __init__(self, M, alpha, beta, iter_num=100):
        super(BayesLinearPolynomialRegressionWithEvidence, self).__init__(
            M=M, alpha=alpha, beta=beta)
        self.iter_num = iter_num
        self.isnan = False

    def fit_1(self, X, t):
        phi = self.polynomialize(X)
        #固有値・固有ベクトル計算
        lamb, u = np.linalg.eig(self.beta * phi.T.dot(phi))
        lamb = lamb.real
        #gammaの計算
        self.gamma = np.sum(lamb / (self.alpha + lamb))
        #事後分布の分散
        self.w_s = np.linalg.inv(self.alpha * np.eye(phi.shape[1]) +
                                 self.beta * phi.T.dot(phi))
        #事後分布の平均
        self.w_m = self.beta * self.w_s.dot(phi.T).dot(t)
        #alphaの方程式
        self.alpha = self.gamma / (self.w_m.T.dot(self.w_m))
        N = t.shape[0]
        #betaの方程式
        self.beta = (N - self.gamma) / (np.sum(
            np.square(t - phi.dot(self.w_m))))
        if np.isinf(self.alpha) or np.isnan(self.alpha) or np.isinf(self.beta):
            if not self.isnan:
                self.isnan = True
                print("[W] alpha or beta value is inf")
            return
        #E(m_n)の計算
        Emn = 0.5*self.beta*np.sum(np.square(t - phi.dot(self.w_m))) +\
              0.5*self.alpha*self.w_m.T.dot(self.w_m)
        #行列Aの計算
        A = self.alpha * np.eye(self.w_m.shape[0]) + self.beta * phi.T.dot(phi)
        detA = np.linalg.det(A)
        if np.isinf(detA):
            if not self.isnan:
                self.isnan = True
                print("[W] |A| is inf")
            return
        #対数エビデンスの計算
        self.evidence = 0.5*self.M*np.log(self.alpha)+0.5*N*np.log(self.beta)-\
                            Emn-0.5*np.log(detA)-\
                            0.5*N*np.log(2*np.pi)

    def fit(self, X, t):
        #繰り返し計算の回数
        for i in range(self.iter_num):
            self.fit_1(X, t)
            if self.isnan:
                print("[W] fit may failed")
                return
```

```python
X_train, t_train = generate_data(f, 0.25, 100)

fig = plt.figure(figsize=(15, 5))
M = [1, 5, 10]

for i in range(3):
    model = BayesLinearPolynomialRegressionWithEvidence(M[i], 1, 1, 100)
    model.fit(X_train, t_train)

    display_predict_with_lines(
        f,
        X_train,
        t_train,
        model,
        subindex=i+1,
        text=r'$M={:d},\alpha={:.3f},\beta={:.3f},\log E={:.2f}$'.format(
            model.M, model.alpha, model.beta, model.evidence))
```
